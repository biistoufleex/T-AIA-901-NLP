```python
import spacy
from spacy.scorer import Scorer
from spacy.training import Example

import matplotlib.pyplot as plt

def evaluate_model(nlp, test_data):
    examples = []
    for text, annot in test_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annot)
        examples.append(example)
    scores = nlp.evaluate(examples)
    return scores

# Charger le modèle
nlp = spacy.load("../../model/model_sur_entrainer")

results = evaluate_model(nlp, TRAIN_DATA)



# Exemple de données d'évaluation
precision = results['ents_p']
recall = results['ents_r']
f1_score = results['ents_f']

# Noms des métriques
metrics = ['Precision', 'Recall', 'F1 Score']

# Valeurs des métriques
values = [precision, recall, f1_score]

plt.figure(figsize=(10, 6))
plt.bar(metrics, values, color=['blue', 'orange', 'green'])

plt.xlabel('Metrics')
plt.ylabel('Percentage')
plt.title('NLP Model Evaluation')
plt.ylim(0, 100)  # Assuming the values are percentages
for i, value in enumerate(values):
    plt.text(i, value + 1, f"{value:.2f}%", ha='center')

plt.show()


print("Précision:", results["ents_p"])
print("Rappel:", results["ents_r"])
print("Score F1:", results["ents_f"])
```
